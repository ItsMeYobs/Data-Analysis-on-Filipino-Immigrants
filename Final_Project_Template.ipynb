## Indicate your Question #1 here

```python
## Start your visualization codes here
#TOP 5 COUNTRIES THAT FILIPINO IMMIGRANTS LOVES

top_countries = dataset.copy()
top_countries['TOTAL'] = top_countries['TOTAL'].str.replace(',', '').astype(float)

# Exclude 'GRAND TOTAL'
top_countries = top_countries[top_countries['COUNTRY'] != 'GRAND TOTAL']

# Sort by 'TOTAL' column in descending order
top_countries = top_countries.sort_values('TOTAL', ascending=False).head(5)

# Increase the size of the graph
plt.figure(figsize=(10, 6))

plt.bar(top_countries['COUNTRY'], top_countries['TOTAL'])
plt.xlabel('COUNTRIES')
plt.ylabel('TOTAL')
plt.title('Top 5 Countries with Highest Total')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust spacing to prevent label overlapping

# Add the unit of measurement to y-axis label
plt.ylabel('Total (Million)')

plt.show()
```

THE TOP 5 COUNTRY THAT FILIPINO IMMIGRANTS WANTS IS U.S.A, CANADA, JAPAN, AUSTRALIA, ITALY, AND WE ALL KNOW THAT THESE COUNTRY SHOWED IN THE VISUALIZATION ARE COUNTRIES THAT HAVE BETTER ECONOMY

```python
#TOP 5 PURPOSE OF FILIPINO IMMIGRANTS
top_purpose = dataset2.copy()
top_purpose['TOTAL'] = top_purpose['TOTAL'].str.replace(',', '').astype(float)

# Exclude 'GRAND TOTAL'
top_purpose = top_purpose[top_purpose['MAJOR OCCUPATION GROUP'] != 'TOTAL']

# Sort by 'TOTAL' column in descending order
top_purpose = top_purpose.sort_values('TOTAL', ascending=False).head(5)

plt.figure(figsize=(10, 6))

plt.bar(top_purpose['MAJOR OCCUPATION GROUP'], top_purpose['TOTAL'])
plt.xlabel('MAJOR OCCUPATION GROUP')
plt.ylabel('TOTAL')
plt.title('Top 5 Purpose with Highest Total')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust spacing to prevent label overlapping

# Add the unit of measurement to y-axis label
plt.ylabel('Total (Million)')

plt.show()
```

WE CAN SEE THAT THE MAJOR PURPOSE OF FILIPINO IMMIGRANTS IS STUDY ABROAD. WE CAN SAY THAT IMMIGRANTS STUDENTS AIMING FOR BETTER KNOWLEDG AND CAREER.

```python
#AGE GROUP OF FILIPINO IMMIGRANTS
age_group= ["14 - Below","15 - 19","20 - 24","25 - 29","30 - 34","35 - 39","40 - 44","45 - 49","50 - 54","55 - 59","60 - 64","65 - 69","70 - Above","Not Reported /No Response"]
year = dataset.columns[1:]

df = pd.DataFrame()

for group_age in age_group:
    age_data = dataset3[dataset3["AGE GROUP"] == group_age]
    data_age = age_data.iloc[0].values[1:]
    df[group_age] = data_age

df.index = year
df = df.replace(',', '', regex=True).apply(pd.to_numeric, errors='coerce')
df.plot.line()
plt.xlabel("Year")
plt.ylabel("Number of Immigrants")
plt.title("Age Group of")
plt.legend()
plt.show()
```

THE MAJOR AGE GROUP OF FILIPINO IMMIGRANTS IS 14 AND BELOW, WHICH MEANS THE AGE OF HIGH SCHOOL STUDENTS. THIS VISUALIZATION CAN SUPPORT THE ANSWER IN QUESTION NUMBER 3 "WHAT IS THE MAJOR PURPOSE OF FILIPINO IMMIGRANTS".

```python
#GENDER OF MOST ACCEPTED FILIPINO IMMIGRANT
# Filter the DataFrame to include only the "TOTAL" row
total_row = dataset1[dataset1['YEAR'] == 'TOTAL']

# Filter out rows with None values in "MALE" or "FEMALE" column
total_row = total_row.dropna(subset=['MALE', 'FEMALE'])

# Remove commas from "MALE" and "FEMALE" columns and convert to float
male_values = total_row['MALE'].str.replace(',', '').astype(float)
female_values = total_row['FEMALE'].str.replace(',', '').astype(float)

# Extract the "YEAR" column from the total row
years = total_row['YEAR']

# Create the bar graph
plt.bar(years, male_values, label='Male')
plt.bar(years, female_values, label='Female', bottom=male_values)

# Add labels and title to the graph
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Male and Female Population Over the Years')

# Set the maximum limit of the y-axis to 2 million
plt.ylim(0, 1501870)

# Add a legend
plt.legend()

# Display the graph
plt.show()
```

THE MOST ACCEPTED GENDER/SEX IS FEMALE. PROBABLY, MOSTLY OF THE IMMIGRANTS IS STUDENTS, FEMALE AGING 14 AND BELOW

```python
#RICH COUNTRIES
countries = ["UNITED STATES OF AMERICA", "CHINA (P.R.O.C.)", "JAPAN", "GERMANY","CANADA","ITALY"]
year = dataset.columns[1:]

df = pd.DataFrame()

for country in countries:
    country_data = dataset[dataset["COUNTRY"] == country]
    country_immigrants = country_data.iloc[0].values[1:]
    df[country] = country_immigrants

df.index = year
df = df.replace(',', '', regex=True).apply(pd.to_numeric, errors='coerce')
df.plot.line()
plt.xlabel("Year")
plt.ylabel("Number of Immigrants")
plt.title("Number of Immigrants in Rich Countries (according to Google)")
plt.legend()
plt.show()
```

USA, JAPAN, CANADA, ITALY ARE COUNTRIES THAT HAVE A BETTER ECONOMY. WE CAN SAY THAT THE ECONOMY OF THE COUNTRY WILL AFFECT THE DECISION OF FILIPINO IMMIGRANT. THIS VISUALIZATION CAN ALSO CONCLUDE THE QUESTION NUMBER 1.

```python
#HAPPY COUNTRIES
countries = ["FINLAND", "DENMARK", "NORWAY", "SWITZERLAND", "ICELAND", "NETHERLANDS", "SWEDEN"]
year = dataset.columns[1:]

df = pd.DataFrame()

for country in countries:
    country_data = dataset[dataset["COUNTRY"] == country]
    country_immigrants = country_data.iloc[0].values[1:]
    df[country] = country_immigrants

df.index = year
df = df.replace(',', '', regex=True).apply(pd.to_numeric, errors='coerce')

df.plot.line()
plt.xlabel("Year")
plt.ylabel("Number of Immigrants")
plt.title("Number of Immigrants in Happy Countries (according to Google)")
plt.legend()

# Adjust y-axis ticks
plt.yticks(np.arange(0, df.max().max()+1, 1000))

plt.show()
```

HAPPY COUNTRIES CAN ALSO ATTRACT FILIPINO TO MIGRATE TO THEIR COUNTRY TO LIVE FOR GOOD.

```python
#SAD COUNTRIES
countries = ["LESOTHO", "YEMEN", "SYRIA"]
year = dataset.columns[1:]

df = pd.DataFrame()

for country in countries:
    country_data = dataset[dataset["COUNTRY"] == country]
    country_immigrants = country_data.iloc[0].values[1:]
    df[country] = country_immigrants

df.index = year
df = df.replace(',', '', regex=True).apply(pd.to_numeric, errors='coerce')
df.plot.line()
plt.xlabel("Year")
plt.ylabel("Number of Immigrants")
plt.title("Number of Immigrants in Sad Countries (according to Google)")
plt.legend()
plt.show()
```
